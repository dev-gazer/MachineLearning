#**01. Importing libraries**
Once again, the needed libraries follow:

*   Numpy - Perform mathematical operations;
*   Pandas - Perform data manipulation and analysis;
*   Matplotlib - Library for plotting and visualizing data.



```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# **02. Loading data from a .csv file**
Let us load the Position_Salaries.csv file and allocate their columns to vectors X and Y, representative of X and Y axis of a chart. We will use the *.head() to show the first five entries of the table.

```
dataset= pd.read_csv('/content/drive/My Drive/Colab Notebooks/Study Datasets/Position_Salaries.csv')
X= dataset.iloc[:,1:2].values
Y= dataset.iloc[:,2].values
dataset
```

# **03. Fitting the random forest model to the dataset**
This step is performed in order to apply the random fores regression. For this, we'll be using the sklearn library ***sklearn.ensemble***.

```
from sklearn.ensemble import RandomForestRegressor
regressor= RandomForestRegressor(n_estimators=300,random_state=0)
regressor.fit(X,Y)
```

#**04. Visualizing the random forest results**
The results for the random forest follow:

```
X_grid=np.arange(min(X),max(X),0.01)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color='red', label='Dataset')
plt.plot(X_grid,regressor.predict(X_grid),color = 'blue', label='Random Forest Fit')
plt.title('Position Level x Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.grid(which='major')
plt.grid(which='minor')
plt.legend(loc='lower right')
plt.show()
```

#**05. Predicting a new result with the random forest regression model**
We'll then investigate if the prediction leads to a salary comparable to the candidate expectation. For a 6.5 level, it follows:

```
regressor.predict([[6.5]])
```

#**06. Conclusion**
The candidate salary expectations are within the range of the company payroll.
